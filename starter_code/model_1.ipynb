```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.

selected_features = df[['koi_disposition', 'koi_period', 'koi_time0bk', 'koi_impact', 'koi_duration', 'koi_depth', 
                        'koi_prad', 'koi_teq', 'koi_insol', 'koi_model_snr', 'koi_steff', 'koi_slogg', 'koi_srad', 
                        'ra', 'dec', 'koi_kepmag']]
selected_features
```

```python
X=selected_features[['koi_period', 'koi_time0bk', 'koi_impact', 'koi_duration', 'koi_depth', 
                        'koi_prad', 'koi_teq', 'koi_insol', 'koi_model_snr', 'koi_steff', 'koi_slogg', 'koi_srad', 
                        'ra', 'dec', 'koi_kepmag']]
y=selected_features['koi_disposition']
y
```

```python
# Dependent variable seems to be string.  
# to get unique values from list 
# reference https://www.geeksforgeeks.org/python-get-unique-values-list/
  
# function to get unique values 
def unique(list1): 
  
    # intilize a null list 
    unique_list = [] 
      
    # traverse for all elements 
    for x in list1: 
        # check if exists in unique_list or not 
        if x not in unique_list: 
            unique_list.append(x) 
    # print list 
    for x in unique_list: 
        print(x)


unique(y)
```

```python
# If 'CANDIDATE', we don't know whether this is a planet yet. For now, drop "candidate' rows.
selected_features1 = selected_features[selected_features.koi_disposition != 'CANDIDATE']
selected_features1
```

```python
# Let's change CONFIRME to 1, FALSE POSITIVE to 0;
# Add a new column named 'koi_disposition1' 

selected_features1['koi_disposition1']=[1 if x =='CONFIRMED' else 0 for x in selected_features1['koi_disposition']] 

print(selected_features1)
```

```python
# recreate X and y based on selected_features1

X=selected_features1[['koi_period', 'koi_time0bk', 'koi_impact', 'koi_duration', 'koi_depth', 
                        'koi_prad', 'koi_teq', 'koi_insol', 'koi_model_snr', 'koi_steff', 'koi_slogg', 'koi_srad', 
                        'ra', 'dec', 'koi_kepmag']]
y=selected_features1['koi_disposition1']
y
```

```python
# Reshape the data, so scaler can run later
X=selected_features1[['koi_period', 'koi_time0bk', 'koi_impact', 'koi_duration', 'koi_depth', 
                        'koi_prad', 'koi_teq', 'koi_insol', 'koi_model_snr', 'koi_steff', 'koi_slogg', 'koi_srad', 
                        'ra', 'dec', 'koi_kepmag']]
y = selected_features1["koi_disposition1"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42)
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler

X_scaler = MinMaxScaler().fit(X_train)
y_scaler = MinMaxScaler().fit(y_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

```python
# Create a Logistic Regression Model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train_scaled, y_train_scaled)
```

# Train the Model



```python
print(f"Training Data Score : {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```
